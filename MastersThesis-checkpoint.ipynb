# Code used for MSc Thesis Project "Looking at HIV-1 Transmission in Quebec Through a Phylogenetic Lens"

```python
#First cell to import all the packages
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt 
import seaborn as sns
from statannot import add_stat_annotation
from pandas.compat import StringIO
from scipy import stats
```

```python
#importing the dataset from SAS and selecting only the years 2002 - 2017
SASHiv = pd.read_excel('RISKGROUPS.xlsx')
clinical = pd.read_excel('CLINICAL.xlsx')
mask = (SASHiv['YearInfection'] >= 2002) & (SASHiv['YearInfection'] <= 2017) #mask that allows set up the year condition
goodyears = SASHiv.loc[mask]
```

## Figure 6 - Timeline of MSM HIV-1 Epidemic in Quebec

```python

#Creating a pivot table for the year data
yeargroup = goodyears.groupby(['YearInfection','Class']).size().to_frame('count').reset_index() #makes a new count column for all years
pivotyear = yeargroup.pivot(index = 'YearInfection', columns = 'Class', values = 'count')


pos = list(range(len(pivotyear['Large Cluster']))) #creates a list of X positions
width = 0.25

#plotting the bars
fig, ax = plt.subplots(figsize =(10,5))
#order is singleton, large cluster and then small cluster
#create a bar with Singletons data, in position pos
plt.bar(pos,
       pivotyear['Singleton'],
       width,
       alpha = 0.5,
       color = '#A30000') #gotta use a way to change the labels to the inde

#create a bar with Large Clusters data in position post + with some width buffer
yeartags = [2002,2003,2004,2005,2006,2007,2008,2009,2010,2011,2012,2013,2014,2015,2016,2017]
plt.bar([p + width for p in pos],
       pivotyear['Large Cluster'],
       width,
       alpha = 0.5,
       color = '#004777') 
#create a bar with Small Cluster data in position post + with some width buffer
plt.bar([p + width*2 for p in pos],
       pivotyear['Small Cluster'],
       width,
       alpha = 0.5,
       color = '#00AFB5') 
#y axis label
ax.set_ylabel('Count')

#Set the chart's title
ax.set_title('Expansion of HIV-1 Clusters through Years (2002-2017)')

#set the position of the x ticks
ax.set_xticks([p + 1.5 * width for p in pos])
#labels for x ticks
ax.set_xticklabels(yeartags)
#set x-axis and y-axix limits
plt.xlim(min(pos)-width, max(pos)+width*4)
plt.ylim([0, max(pivotyear['Singleton'])] )

plt.legend(['Singletons', 'Large Clusters', 'Small Cluster'], loc='best', )
```

## Figure 7 - Comparing Recency and Viral Load between Classes with Boxplots

```python
#boxplot for PctDiv
plt.figure(figsize=(9,10))
sns.set(style = "whitegrid")
ax = sns.boxplot(x="Class",y="Pct_Div",data=goodyears, 
            showfliers=False, linewidth=1.5,
            dodge = False)
order = ['Singleton', 'Large Cluster', 'Small Cluster'] #order of appeareance

pctbox = add_stat_annotation(ax, data=goodyears, x="Class", y="Pct_Div",
                   box_pairs=[("Singleton","Large Cluster"), ("Singleton","Small Cluster"), ("Large Cluster", "Small Cluster")],
                    test='Mann-Whitney',text_format='star', loc='outside', verbose=2)
```

```python
#boxplot for logVL
ax = sns.boxplot(x="Class",y="logVL",data=goodyears, 
            showfliers=False, linewidth=1.5,
            dodge = False)

order = ['Singleton', 'Large Cluster', 'Small Cluster']

pctbox = add_stat_annotation(ax, data=goodyears, x="Class", y="logVL", order=order,
                   box_pairs=[("Singleton","Large Cluster"), ("Singleton","Small Cluster"), ("Large Cluster", "Small Cluster")],
                    test='Mann-Whitney',text_format='star', loc='outside', verbose=2)
```

## Figures 8 and 9 - Age Distribution, Histograms and Kernel Density Plots

### Figure 9 Age Distribution

```python
#This step creates a new datafram that contains more attributes than the previous goodyears dataframe
goodyearmask = goodyears[['Lab_ID','Class','Patient_ID','Database_ID','sex','Region','DateVisit','NewAge','YearInfection','Cluster_ID','Pct_Div']]
clinicalmask = clinical[['State','AgeGroup','Period','Clinical','Lab_ID']]
merged_inner = pd.merge(left=goodyearmask,right=clinicalmask, left_on='Lab_ID', right_on='Lab_ID')

#Check age distribution by Class Group
yearagestd = merged_inner.groupby(['YearInfection','Class'])['NewAge'].std()
plotclassstd = merged_inner.groupby(['YearInfection','Class'])['NewAge'].std()

LCage = merged_inner.loc[(merged_inner['Class']=='Large Cluster')]
SCage = merged_inner.loc[(merged_inner['Class']=='Small Cluster')]
Sage = merged_inner.loc[(merged_inner['Class']=='Singleton')]


sns.set_style('darkgrid')
big  = sns.distplot(merged_inner['NewAge'], bins='auto', norm_hist=True, kde=False)
p1 = sns.distplot(LCage.NewAge,hist = False, kde=True, color = 'blue', label = 'Large Cluster')
p2 = sns.distplot(SCage.NewAge,hist = False, kde=True, color = 'orange', label = 'Small Cluster')
p3 = sns.distplot(Sage.NewAge, hist = False, kde=True, color = 'green', label = 'Singleton')
plt.legend()
```

### Figure 8 ECDF Curve 

```python
#Creating the ECDF curve
totalage = np.array(merged_inner.NewAge)
Larray=np.array(LCage.NewAge)
Sarray = np.array(SCage.NewAge)
sinarray = np.array(Sage.NewAge)
#Defining ECDF Function
def ecdf(data):
    n=len(data)
    x=np.sort(data)
    y=np.arange(1, n+1)/n
    return x,y

#applying ECDF function
np.sort(Larray)
x,y = ecdf(totalage)
xl, yl = ecdf(Larray)
xs, ys = ecdf(Sarray)
xsin, ysin = ecdf(sinarray)

# Specify array of percentiles: percentiles
percentiles = np.array([25, 50, 75])

#percentiles on each array
Ltiles = np.percentile(Larray, percentiles)
sintiles = np.percentile(sinarray, percentiles)
Stiles = np.percentile(Sarray, percentiles)

#Defining colours for each Class Group
_ = plt.plot(xl, yl, color='blue', label='Large Cluster')
_ = plt.plot(xs, ys,  color='orange', label='Small Cluster')
_ = plt.plot(xsin, ysin, color='green', label='Singleton')

#position of the diamonds on the chart, first one 25 percentile, then median, finally 95 percentile
_ = plt.plot(Stiles, percentiles/100, marker='.', color='blue',
         linestyle='none')
_ = plt.legend(loc='lower right')
_ = plt.xlabel('Age')
_ = plt.ylabel('Percentiles')
print("Sample total Number: ",len(totalage))
print("Sample total Large Clusters: ",len(Larray))
print("Sample total Small Clusters: ",len(Sarray))
print("Sample total Singletons: ",len(sinarray))
print("Median age Large Clusters: ",np.median(Larray))
print("Median age Singletons: ",np.median(sinarray))
```

## Figure 13 Determining Viral Tropism of LC, SC and Singleton Viruses

```python

tropism = pd.read_excel('rtsimple2.xlsx', index_col=2) #read the data
tropism[["R5_RT", "X4_RT"]] = tropism[["R5_RT", "X4_RT"]].astype(str).astype(int) #convert R4 and X4 values to string and then to integers (previously objects)
tropism[["Cluster_Type","Cluster_ID"]]=tropism[["Cluster_Type","Cluster_ID"]].astype(str)
tpnumbers = tropism[["R5_RT", "X4_RT"]] #Not needed
tags = pd.read_excel('tagstropism.xlsx', index_col=0) #tags for the cluster type

#Slice the table only for R5 and X4 values, it's important to convert the objects to numbers and then obtain their log10 values
tpnumbers2= tropism[["R5_RT", "X4_RT"]]
tags2 =  pd.read_excel('tagstropism2.xlsx', index_col=0)
avg = tropism.groupby('Sample_ID')["R5_RT", "X4_RT"].mean()
joined2 = np.log10(avg)
logtags2 = joined2.join(tags2) #add NSI column to tags2 spreadsheet to show that growth 

#Creating a color palete with the same number of colors as unique values in the Cluster_Type column
flatui = ["#006400", "#ffffff", "#1e90ff", "#ff8c00"]
cluster_pal = sns.color_palette(flatui,len(logtags2.Cluster_Type.unique()))
#Create a dictionary where the key is the category and the values are the
#colors from the palette we just created
cluster_lut = network_lut = dict(zip(logtags2.Cluster_Type.unique(), cluster_pal))
#get the series of all of the categories
clusters = logtags2.Cluster_Type
#map the colors to the series. Now we have a list of colors the same
#length as our dataframe, where unique values are mapped to the same color
cluster_colors = pd.Series(clusters).map(cluster_lut)

#colours for NSI Col
flatui2 = ["#00008b","#8b0000", "#ffffff"]
MT2_pal = sns.color_palette(flatui2,len(logtags2.MT2.unique()))
#Create a dictionary where the key is the category and the values are the
#colors from the palette we just created
MT2_lut = network_lut = dict(zip(logtags2.MT2.unique(), MT2_pal))
#get the series of all of the categories
MT2 = logtags2.MT2
#map the colors to the series. Now we have a list of colors the same
#length as our dataframe, where unique values are mapped to the same color
MT2_colors = pd.Series(MT2).map(MT2_lut)


plt.figure(figsize=(40,30))
sns.clustermap(logtags2[["R5_RT", "X4_RT"]], standard_scale=1, cmap='mako', 
               method='single',robust=True, col_cluster=False,
               metric='euclidean', row_colors=[cluster_colors,MT2_colors], 
               annot=True, annot_kws={"size": 7})

```
